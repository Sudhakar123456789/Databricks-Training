step 1: create a dataframe from this path: /Volumes/acc_dev/naval/raw/financial_dataset.csv
Step 2: rename the all cols and make it lower case
        ii. add new col for auditing: (ingestion date)
        iii. add new col for path: (input_path)
Step 3: save final df to table       
note: keep it clean Modular code

```python
from pyspark.sql.functions import *
```

```python
df=spark.read.csv("/Volumes/acc_dev/naval/raw/financial_dataset.csv",inferSchema=True,header=True)
```

```python
df.display()
```

```python
a = [col.lower() for col in df.columns]
```

```python
df=df.toDF(*a)
```

```python
df.display()
```

```python
df=df.withColumn("ingestion_date",current_timestamp()).withColumn("path",col("_metadata.file_path"))
```

```python
df.display()
```

```python
df.write.saveAsTable("acc_dev.sudhakar.financial")
```
